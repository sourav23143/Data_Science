```python
df1 = pd.DataFrame({'key1':[1, 2, 4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'key1':[1, 2, 45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]})
```

```python
df2
```

```python
#merge>>
```

```python
pd.merge(df1, df2, how = 'inner')  #it will show common element
```

```python
#show common elemnts of key1 in both and  and corrsponding elements of 1, 2, 3 in key2, key3, ,key4, key5
```

```python
pd.merge(df1, df2, how = 'left') #all elements of df1 and and correcspoding elements to key1 in df2
```

```python
pd.merge(df1, df2, how = 'right') #all elements of df2 and and correcspoding elements to key1 in df1
```

```python
pd.merge(df1, df2, how = 'outer')  #gives all the keys from both dataframe
```

```python
df1 = pd.DataFrame({'key1':[1, 2, 4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]},
                  index = ['a','b','c','d','e'])
```

```python
df2 = pd.DataFrame({'key6':[1, 2, 45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]},
                   index = ['a','b','h','i','j'])
```

```python
df1
```

```python
df2
```

```python
#here in both dataframe we can see index "a" and "b" are only common index  and all keys name are different
```

```python
#in join parameter is not keys but index
```

```python
df1.join(df2, how= "inner")  #Join columns of another DataFrame.
```

```python
#so in inner join we get what ever common index values in df1 and df2 
```

```python
df1.join(df2, how= "left") 
```

```python
#all the elements from left and corresponding elements from common index from df2 
```

```python
df1.join(df2, how= "outer") 
```

```python
#here we get all the elements from both >> i.e all the indexs from df1 and df2 and matching values and Na for which not matches
```

```python
df1.join(df2, how= "cross") 
```

```python
#here we get all the permutation and combination
```
